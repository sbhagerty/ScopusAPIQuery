# Setting up Scopus API Query to be easily updated

```python
import requests
import pandas as pd
import config
import ScopusFunctions
```

Took a list of names and added columns with authorID and TotalPapers in Scopus 

```python
for index, person in FacultyList.iterrows():
    authID, docCount = ScopusFunctions.getAuthIDFromName(person['FirstName'], person['LastName'])
    FacultyList.loc[index,'authID'] = authID
    FacultyList.loc[index,'TotalPapers']=docCount
```

I saved that dataframe to a csv. Then went and manually checked and input authIDs for anyone who did not get a match in previous step.  

```python
FacultyList_withIDs=pd.read_csv('ListOfFaculty.csv')
FacultyList_withIDs.head()
```

Now I'm going to get totalDocs for anyone who didn't have author ID's before, these people would have a 0 value for doc count. 

```python
for index, person in FacultyList_withIDs.iterrows():
    if person['TotalPapers'] == 0:
        print(person['authID'])
        query ='AU-ID('+str(person['authID'])+')'
        query_params = {'query':query}
        baseurl = 'http://api.elsevier.com/content/search/author'
        response = requests.get(baseurl, headers=config.apiKey, params = query_params)
        data = response.json()
        resultCount = data['search-results']['entry'][0]['document-count']
        FacultyList_withIDs.loc[index, 'TotalPapers'] = resultCount
```
