Okay so first I need to combine all the papers I have into one big dataframe.  I had set up the initial query by department, so all the department papers need to be combined now. 

```python
math = pd.read_csv('BrandeisMath/MathDeptPapers_04032019.csv')
chem = pd.read_csv('BrandeisChemistry/ChemistryPapers.csv')
bio = pd.read_csv('BrandeisBiology/BiologyPapers.csv')
biochem = pd.read_csv('BrandeisBiochemistry/BiochemistryPapers.csv')
compSci = pd.read_csv('BrandeisComputerScience/ComputerSciencePapers.csv')
```

```python
depts = [math, chem, bio, biochem, compSci]
AllPapers=pd.concat(depts)
AllPapers.columns
AllPapers= AllPapers.drop('Unnamed: 0', axis=1)
```

```python
AllPapers.to_csv('AllPapers.csv') #save the dataframe. 
```
